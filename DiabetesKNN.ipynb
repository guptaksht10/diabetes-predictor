```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
diabetes_dataset = pd.read_csv('./data/diabetes.csv')
```

```python
x_data, y_data = diabetes_dataset.drop(columns=['Outcome']), diabetes_dataset['Outcome']
```

```python
x_data
```

```python
y_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 2)
```

```python
scale = preprocessing.StandardScaler()

x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

```python
clf = KNeighborsClassifier(5, algorithm = 'brute')
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
k_values = []
accuracy = []
for param in range(1, 51) :
    clf = KNeighborsClassifier(param, algorithm = 'brute')
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    accuracy.append(accuracy_score(y_test, y_pred))
    k_values.append(param)

plt.figure(figsize = (10, 7))
plt.plot(k_values,accuracy, marker = 'o')
plt.xlabel('k_values')
plt.ylabel('Accuracy')
plt.title("KNN Accuracy vs. k")
plt.show()
```
